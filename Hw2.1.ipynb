```python
import pandas as pd
import numpy as np

tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", 
                   match="Коефіцієнт народжуваності в регіонах України")

tmp[0].head()
```

```python
tmp[0].shape
```

```python
table = tmp[0].replace("—", np.nan)
table
```

```python
table.dtypes
```

```python
table["2014"] = pd.to_numeric(table["2014"])
table["2019"] = pd.to_numeric(table["2019"])
table.dtypes
```

```python
table.isnull().sum()/table.shape[0]
```

```python
table = table.drop([27])
table
```

```python
table = table.fillna({"1950": table["1950"].mean().round(),
                      "1960": table["1960"].mean().round(),
                      "1970": table["1970"].mean().round(), 
                      "1990": table["1990"].mean().round(),
                      "2000": table["2000"].mean().round(), 
                      "2012": table["2012"].mean().round(), 
                      "2014": table["2014"].mean().round(),
                      "2019": table["2019"].mean().round()})
table
```

```python
table[table["2019"] > table["2019"].mean()][["Регіон", "2019"]]
```

```python
table.sort_values(["2014"], ascending=False)[["Регіон", "2014"]].head(1)
```

```python
table1 = table.plot.bar(
    x="Регіон",
    y="2019",
    title="Діаграма народжуваності по регіонах у 2019 році"
)
table1
```

```python
table2 = table.plot.pie(
    x="Регіон",
    y="2019",
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)
table2
```
